```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
first_dataset = pd.read_csv("Unemployment in India.csv")
```

```python
first_dataset.shape
```

```python
first_dataset.isna().sum()
```

```python
first_dataset.isnull().sum()
```

```python
first_dataset.dropna(inplace=True)
```

```python
first_dataset.head()
```

```python
region_cols = first_dataset['Region'].value_counts().index
for col in region_cols:
    df_region = first_dataset[first_dataset['Region'] == col]
    plt.figure(figsize=(12,8))
    sns.barplot(df_region, x='Date', y='Estimated Employed', hue='Area')
    plt.title(col)
    plt.xticks(rotation=90)
    plt.show()
```

```python
for col in region_cols:
    df_region = first_dataset[first_dataset['Region'] == col]
    plt.figure(figsize=(12,8))
    sns.barplot(df_region, x='Date', y='Estimated Labour Participation Rate (%)', hue='Area')
    plt.title(col)
    plt.xticks(rotation=90)
    plt.show()
```
