```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
second_dataset = pd.read_csv('Unemployment_Rate_upto_11_2020.csv')
```

```python
second_dataset.rename(columns={'Region.1':'Zone'}, inplace=True)
second_dataset.head()
```

```python
zone_cols = second_dataset['Zone'].value_counts().index
for col in zone_cols:
    df_zone = second_dataset[second_dataset['Zone'] == col]
    plt.figure(figsize=(12,8))
    sns.barplot(df_zone, x='Date', y='Estimated Employed', hue='Region')
    plt.title(col)
    plt.xticks(rotation=90)
    plt.show()
```

```python
for col in zone_cols:
    df_zone = second_dataset[second_dataset['Zone'] == col]
    plt.figure(figsize=(12,8))
    sns.barplot(df_zone, x='Date', y='Estimated Labour Participation Rate (%)', hue='Region')
    plt.title(col)
    plt.xticks(rotation=90)
    plt.show()
```
